```python
import pandas as pd
```

```python
def read_json_partial(file_path, num_rows=1000):
    return pd.read_json(file_path, lines=True, nrows=num_rows)

def print_head(df, num_rows=5, exclude_column=None):
    if exclude_column and exclude_column in df.columns:
        df = df.drop(columns=[exclude_column])
    print(df.head(num_rows))

def main():
    # Define file paths
    business_path = 'yelp_academic_dataset_business.json'
    checkin_path = 'yelp_academic_dataset_checkin.json'
    review_path = 'yelp_academic_dataset_review.json'
    tip_path = 'yelp_academic_dataset_tip.json'
    user_path = 'yelp_academic_dataset_user.json'

    # Read the first 1000 rows of each file
    print("Processing business data...")
    business_df = read_json_partial(business_path)
    print("processed_business.csv")
    print_head(business_df)
    business_df.to_csv('processed_business.csv', index=False)

    print("\nProcessing checkin data...")
    checkin_df = read_json_partial(checkin_path)
    print("processed_checkin.csv")
    print_head(checkin_df)
    checkin_df.to_csv('processed_checkin.csv', index=False)

    print("\nProcessing review data...")
    review_df = read_json_partial(review_path)
    print("processed_review.csv")
    print_head(review_df, exclude_column='user_id')
    review_df.to_csv('processed_review.csv', index=False)

    print("\nProcessing tip data...")
    tip_df = read_json_partial(tip_path)
    print("processed_tip.csv")
    print_head(tip_df, exclude_column='user_id')
    tip_df.to_csv('processed_tip.csv', index=False)

    print("\nProcessing user data...")
    user_df = read_json_partial(user_path)
    print("processed_user.csv")
    print_head(user_df)
    user_df.to_csv('processed_user.csv', index=False)

    print("\nData processing complete.")

if __name__ == "__main__":
    main()
```
